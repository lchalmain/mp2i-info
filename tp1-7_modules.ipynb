# Créer son propre module

Maintenant que nous avons défini un certain nombre de fonctions, on peut vouloir les réutiliser dans d'autres programmes.<br>
Pour créer son propre module, par exemple le module `liste`, on crée un fichier `liste.ml` (sous VSCodium par exemple).<br>
On pourra par exemple mettre dans ce fichier la fonction `tri_fusion` définie dans un précédent TP :
```ocaml
let rec tri_fusion liste =
    let rec split = function
        | [] -> ([], [])
        | [e] -> ([e], [])
        | e1::e2::q -> let q1, q2 = split q in (e1::q1, e2::q2)
    in
    let rec fusion_triee = function
        | [], [] -> []
        | l1, [] -> l1
        | [], l2 -> l2
        | e1::q1, e2::q2 -> if e1 < e2 then e1::(fusion_triee (q1, e2::q2) )
                            else e2::(fusion_triee (e1::q1, q2) )
    in
    match liste with
        | [] | [_] -> liste (* les listes qu'on ne peut plus couper en deux sont celles de 0 ou 1 élément *)
        | _ -> let l1, l2 = split liste in
                    fusion_triee (tri_fusion l1, tri_fusion l2)
```

Ce fichier doit ensuite être compilé dans le terminal avec la commande `ocamlc -c liste.ml`.<br>
*Remarque : on utilise l'option `-c` car l'édition de liens est inutile ici.*<br>
La fonction `tri_fusion` du module `Liste` pourra alors être appelée dans un autre programme en indiquant `Liste.tri_fusion`.<br>
En mode interactif on commencera par charger le module en indiquant `#load "liste.cmo";;` avant de pouvoir utiliser le module de la même façon.
