 **<p style = "color:purple">Exercice :</p>** 
 
 1. Écrire une fonction `max` qui renvoie le maximum de deux arguments.
 1. Observer la signature de la fonction. Déterminer trois types possibles pour `'a`.

````ocaml
let max x y =
    if x > y then x
    else y
;;
````

````ocaml
assert (max 2 3 = 3);
assert (max 5. (-1.) = 5.);
assert (max "mot" "chaine" = "mot")
````

La fonction `max` est définie pour des arguments appartenants à un ensemble totalement ordonné. Ces arguments peuvent être de type `int`, `float` ou même `string`(ordre lexicographique).

 **<p style = "color:purple">Exercice :</p>** 
 Définir en OCaml la fonction $\tanh : x \longmapsto \dfrac{e^x - e^{-x}}{e^x + e^{-x}}$ en utilisant un seul appel à la fonction `exp : float -> float` de OCaml.

````ocaml
let tanh x = let e = exp x in
    (e -. 1. /. e) /. (e +. 1. /. e)
;;
````

````ocaml
assert (tanh 0. = 0.)
````

### Géométrie

Le type `float*float` désigne un couple de flottants, et peut représenter un point `p` dans $\mathbb{R}^2$.

On peut récupérer les coordonnées de `p` avec la déclaration `let x, y = p` . Par exemple :

````ocaml
let p = (2.1, 3.7) in (* exemple de point *)
let x, y = p in (* récupération des coordonnées *)
x;; (* affichage de x *)
````

 **<p style = "color:purple">Exercice :</p>** 
 1. Écrire une fonction `distance : float*float -> float*float -> float` telle que `distance p1 p2` renvoie la distance euclidienne entre les points `p1` et `p2`.  
1. Vérifier que la distance entre $(0, 0)$ et $(1, 1)$ est (approximativement) $\sqrt{2}$.

````ocaml
let distance p1 p2 = let (x1, y1) = p1 and (x2, y2) = p2 in
    sqrt ( (x2 -. x1)**2. +. (y2 -. y1)**2. )
;;
````

````ocaml
assert ( (distance (0., 0.) (1., 1.)) -. (sqrt 2.) < 0.00001)
````

 **<p style = "color:purple">Exercice :</p>** 
 Écrire une fonction `polaire : float*float -> float*float` qui, étant donné les coordonnées polaires $(r, \theta)$ d'un point, renvoie ses coordonnées cartésiennes $(x, y)$. On rappelle que $x = r\cos(\theta)$ et $y = r\sin(\theta)$. On pourra utiliser `cos` et `sin` en OCaml.

````ocaml
let polaire (r, theta) =
    (r *. (cos theta), r *. (sin theta))
;;
````

En fait, on peut déconstruire le tuple en argument dans la fonction comme dans l'exercice précédent, ou bien comme ici le faire directement dans l'argument.

 **<p style = "color:purple">Exercice :</p>** 
 Écrire une fonction `milieu : float*float -> float*float -> float*float` telle que `milieu p1 p2` renvoie le milieu du segment d'extrémités `p1` et `p2`.

````ocaml
let milieu (x1, y1) (x2, y2) = (x1 +. x2) /. 2. , (y1 +. y2) /. 2.;;
````

````ocaml
assert (milieu (1., 5.) (-3., 7.) = (-1., 6.))
````

 **<p style = "color:purple">Exercice :</p>** 
 Écrire une fonction `orthogonaux : int*int -> int*int -> bool` qui détermine si deux vecteurs sont orthogonaux ou non.

````ocaml
let orthogonaux u1 u2 = let (x1, y1) = u1 and (x2, y2) = u2 in
    x1 *. x2 +. y1 *. y2 = 0.
;;
````

````ocaml
assert (orthogonaux (3., 4.) (4., -3.));
assert (not (orthogonaux (5., -2.) (5., 8.)) );;
````
